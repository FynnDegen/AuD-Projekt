```c++
#include <algoviz/SVG.hpp>
AlgoViz::clear();
SVG canvas = SVG(1700, 1500);
```

```c++
class menu{
    private:
        Rect background;
        Rect buttonRim;
        Rect button1;
        Rect button2;
        Rect button3;
        Text titleText;
        Text button1Text;
        Text button2Text;
        Text button3Text;
        Text diffText;
        Text escText;
        int buttonCounter = 0;
        int menuCounter;
        int diffSetting = 0;
        string lastKey;
        SVG *canvas;
    
    public:
    
        menu(SVG *canvas){
            this -> canvas = canvas;
            mainMenu();
        }
        
        void mainMenu(){
            
            menuCounter = 0;
            buttonCounter = 0;
            background = Rect(0, 0, 5000, 5000, canvas);
            background.setFill("lightBlue");
            buttonRim = Rect(0, 0, 0, 0, canvas);
            titleText = Text("Cyber Crawler", 280, 200, canvas);
            titleText.setAttribute("font-size", 180);
            
            button1 = Rect(360, 360, 980, 180 , canvas);
            button2 = Rect(360, 610, 980, 180, canvas);
            button3 = Rect(360, 860, 980, 180, canvas);
            
            button1Text = Text("Spielen", 650, 480, canvas);
            button1Text.setAttribute("font-size", 100);
            button1Text.setAttribute("fill", "white");
            
            button2Text = Text("Schwierigkeit", 550, 730, canvas);
            button2Text.setAttribute("font-size", 100);
            button2Text.setAttribute("fill", "white");
            
            button3Text = Text("Mitwirkende", 575, 980, canvas);
            button3Text.setAttribute("font-size", 100);
            button3Text.setAttribute("fill", "white");
        }
        
        void diffMenu(){
            
            menuCounter = 1;
            buttonCounter = 0;
            background = Rect(0, 0, 5000, 5000, canvas);
            background.setFill("red");
            buttonRim = Rect(0, 0, 0, 0, canvas);
            titleText = Text("Schwierigkeitsgrad", 100, 200, canvas);
            titleText.setAttribute("font-size", 180);
            
            button1 = Rect(360, 360, 980, 180 , canvas);
            button2 = Rect(360, 610, 980, 180, canvas);
            button3 = Rect(360, 860, 980, 180, canvas);
            
            button1Text = Text("Leicht", 700, 480, canvas);
            button1Text.setAttribute("font-size", 100);
            button1Text.setAttribute("fill", "white");
            
            button2Text = Text("Medium", 675, 730, canvas);
            button2Text.setAttribute("font-size", 100);
            button2Text.setAttribute("fill", "white");
            
            button3Text = Text("Schwer", 675, 980, canvas);
            button3Text.setAttribute("font-size", 100);
            button3Text.setAttribute("fill", "white");
            
            switch(diffSetting){
                    
                case 0:
                    diffText = Text("Der Schwierigkeitsgrad ist auf LEICHT gesetzt.", 50, 1200, canvas);
                    diffText.setAttribute("font-size", 75);
                    break;
                    
                case 1:
                    diffText = Text("Der Schwierigkeitsgrad ist auf MEDIUM gesetzt.", 50, 1200, canvas);
                    diffText.setAttribute("font-size", 75);

                    break;
                    
                case 2:
                    diffText = Text("Der Schwierigkeitsgrad ist auf SCHWER gesetzt.", 50, 1200, canvas);
                    diffText.setAttribute("font-size", 75);
                    break;
                    
                default:
                    break;  
            }
            
            escText = Text("Esc, um zum Hauptmen端 zur端ckzukehren.", 0, 1490, canvas);
            escText.setAttribute("font-size", 50);
        }
    
        void creatorsMenu(){
            menuCounter = 2;
            background = Rect(0, 0, 5000, 5000, canvas);
            background.setFill("green");
            
            titleText = Text("Mitwirkende", 325, 200, canvas);
            titleText.setAttribute("font-size", 180);

            button1Text = Text("Fynn Degen", 560, 480, canvas);
            button1Text.setAttribute("font-size", 100);
            
            button2Text = Text("Jan Miller", 600, 730, canvas);
            button2Text.setAttribute("font-size", 100);
            
            button3Text = Text("Jannis Neuhaus", 460, 980, canvas);
            button3Text.setAttribute("font-size", 100);
            
            escText = Text("ESC, um zum Hauptmen端 zur端ckzukehren.", 0, 1490, canvas);
            escText.setAttribute("font-size", 50);
            
        }
    
        void Update(string lastKey){
        
            if(lastKey == "ArrowDown"){
                if(buttonCounter < 2){
                    buttonCounter++;
                }
                else{
                    buttonCounter = 0;
                }
            } 
            else if(lastKey == "ArrowUp"){
                if(buttonCounter > 0){
                     buttonCounter--;
                }
                else{
                    buttonCounter = 2;
                }
            }   
            
            
            switch(buttonCounter){
                    
                case 0:
                    buttonRim.moveTo(350, 350);
                    buttonRim.setWidth(1000);
                    buttonRim.setHeight(200);
                    buttonRim.setFill("white");
                    break;
                    
                case 1:
                    buttonRim.moveTo(350, 600);
                    buttonRim.setWidth(1000);
                    buttonRim.setHeight(200);
                    buttonRim.setFill("white");
                    break;
                    
                case 2:
                    buttonRim.moveTo(350, 850);
                    buttonRim.setWidth(1000);
                    buttonRim.setHeight(200);
                    buttonRim.setFill("white");
                    break;
                    
                default:
                    break;  
            }
            
            if(lastKey == "Enter" && menuCounter == 0){
                switch(buttonCounter){
                    
                    case 0:
                        //starte Spiel
                        break;
                    
                    case 1:
                        diffMenu();
                        break;
                    
                    case 2:
                        creatorsMenu();
                        break;
                    
                    default:
                        break;  
                }
            }
            else if(lastKey == "Enter" && menuCounter == 1){
                
                switch(buttonCounter){
                    
                    case 0:
                        diffText = Text("Der Schwierigkeitsgrad ist auf LEICHT gesetzt.", 50, 1200, canvas);
                        diffText.setAttribute("font-size", 75);
                        diffSetting = 0;
                        break;
                    
                    case 1:
                        diffText = Text("Der Schwierigkeitsgrad ist auf MEDIUM gesetzt.", 50, 1200, canvas);
                        diffText.setAttribute("font-size", 75);
                        diffSetting = 1;
                        break;
                    
                    case 2:
                        diffText = Text("Der Schwierigkeitsgrad ist auf SCHWER gesetzt.", 50, 1200, canvas);
                        diffText.setAttribute("font-size", 75);
                        diffSetting = 2;
                        break;
                    
                    default:
                        break;  
                }
            }
            
            else if(lastKey == "Escape" && (menuCounter == 1 || menuCounter == 2)){ 
                mainMenu();
            }
        }     
}
```

```c++
menu mainMenu = menu(&canvas);
int i = 1;
string lastKey;
while(i == 1){
    lastKey = canvas.lastKey();
    mainMenu.Update(lastKey);
}
```
