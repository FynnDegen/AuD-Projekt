```c++
class Sword {
    private:
        int dmg;
        int spd;

        int mapPosX;
        int mapPosY;
    
        Image sword;
    
        Enemy enemy[4];
        
    
    public:
        Sword(int mapPosX, int mapPosY) {
            sword = Image("bild.png", test[mapPosY][mapPosX].getX(), test[mapPosY][mapPosX].getY(), 100, 100, &canvas);
            sword.hide();
            enemy[0] = Enemy(100, 5,1,4,3,1,nullptr,nullptr,nullptr,nullptr);
        }
    
        void melee() {
            sword.show();
            AlgoViz::sleep(300);
            sword.hide();
        }
    
        void moveSword(int pMapPosX, int pMapPosY, int pDirec) {
            switch(pDirec) {
                case 0:
                    sword.moveTo(test[pMapPosY - 1][pMapPosX].getX() - 50, test[pMapPosY - 1][pMapPosX].getY() - 50);
                    for(int i = 0; i<4; i++) {
                        if (enemy[i].getMapPosX() == pMapPosX - 1 && enemy[i].getMapPosY() == pMapPosY) {
                            enemy[i].fickerio();
                            
                        }
                    }
                    break;
                case 1:
                    sword.moveTo(test[pMapPosY][pMapPosX - 1].getX() - 50, test[pMapPosY][pMapPosX - 1].getY() - 50);
                    break;
                case 2:
                    sword.moveTo(test[pMapPosY + 1][pMapPosX].getX() - 50, test[pMapPosY + 1][pMapPosX].getY() - 50);
                    break;
                case 3:
                    sword.moveTo(test[pMapPosY][pMapPosX + 1].getX() - 50, test[pMapPosY][pMapPosX + 1].getY() - 50);
                    break;
                default:
                    break;
            }
        }
};
```

```c++
class Player {
    private:
        int hp;
        int attDmg;
        int attSpd;
    
        int x;
        int y;
        int direc;
        int mapPosX;
        int mapPosY;
    
        Image playerUp;
        Image playerLeft;
        Image playerDown;
        Image playerRight;
        Sword sword = Sword(12, 2);
        Arrow arrow = Arrow(1, 10);
    
    public:
        Player(int pMapX, int pMapY) {
            mapPosX = pMapX;
            mapPosY = pMapY;
            playerUp = Image("PlayerUp.png", test[mapPosY][mapPosX].getX(), test[mapPosY][mapPosX].getY(), 100, 100, &canvas);
            playerLeft = Image("PlayerLeft.png", test[mapPosY][mapPosX].getX(), test[mapPosY][mapPosX].getY(),100,100, &canvas);
            playerDown = Image("PlayerDownTest.png", test[mapPosY][mapPosX].getX(), test[mapPosY][mapPosX].getY(), 100, 100, &canvas);
            playerRight = Image("PlayerRight.png", test[mapPosY][mapPosX].getX(), test[mapPosY][mapPosX].getY(), 100, 100, &canvas);
        }
    
        void movePlayerX() {
            setX(test[mapPosX][mapPosY].getMapX());
            movePlayerSprite();
        }
    
        void movePlayerY() {
            setY(test[mapPosX][mapPosY].getMapY());
            movePlayerSprite();
        }
    
        void movePlayerSprite() {
            playerUp.moveTo(test[mapPosY][mapPosX].getX(), test[mapPosY][mapPosX].getY());
            playerLeft.moveTo(test[mapPosY][mapPosX].getX(), test[mapPosY][mapPosX].getY());
            playerDown.moveTo(test[mapPosY][mapPosX].getX(), test[mapPosY][mapPosX].getY());
            playerRight.moveTo(test[mapPosY][mapPosX].getX(), test[mapPosY][mapPosX].getY());
        }

        void keyMovement(std::string pKey) {
            if(pKey == "w") {
                playerUp.toFront();
                direc = 0;
                if(test[mapPosY - 1][mapPosX].getTag() == 0){
                    mapPosY--;
                    movePlayerY();
                    sword.moveSword(mapPosX, mapPosY, 0);
                } else {
                    sword.moveSword(mapPosX, mapPosY, 0);
                }
            } else if(pKey == "a") {
                playerLeft.toFront();
                direc = 1;
                if(test[mapPosY][mapPosX - 1].getTag() == 0){
                    mapPosX--;
                    movePlayerX();
                    sword.moveSword(mapPosX, mapPosY, 1);
                } else {
                    sword.moveSword(mapPosX, mapPosY, 1);
                }
            } else if(pKey == "s") {
                playerDown.toFront();
                direc = 2;
                if(test[mapPosY + 1][mapPosX].getTag() == 0){
                    mapPosY++;
                    movePlayerY();
                    sword.moveSword(mapPosX, mapPosY, 2);
                } else {
                    sword.moveSword(mapPosX, mapPosY, 2);
                }
            } else if(pKey == "d") {
                playerRight.toFront();
                direc = 3;
                if(test[mapPosY][mapPosX + 1].getTag() == 0){
                    mapPosX++;
                    movePlayerX();
                    sword.moveSword(mapPosX, mapPosY, 3);
                } else {
                    sword.moveSword(mapPosX, mapPosY, 3);
                }
            } else if(pKey == "e") {
                arrow.shoot(getX(), getY());
            } else if(pKey == "q") {
                sword.melee();
            } else if(pKey == "ArrowUp") {
                playerUp.toFront();
                direc = 0;
                sword.moveSword(mapPosX, mapPosY, 0);
            } else if(pKey == "ArrowLeft") {
                playerLeft.toFront();
                direc = 1;
                sword.moveSword(mapPosX, mapPosY, 1);
            } else if(pKey == "ArrowDown") {
                playerDown.toFront();
                direc = 2;
                sword.moveSword(mapPosX, mapPosY, 2);
            } else if(pKey == "ArrowRight") {
                playerRight.toFront();
                direc = 3;
                sword.moveSword(mapPosX, mapPosY, 3);
            }
        }
    
        int getHP() {
            return hp;
        }

        void setHP(int pHP) {
            hp = pHP;
        }

        int getAttDmg() {
            return attDmg;
        }

        void setAttDmg(int pAttDmg) {
            attDmg = pAttDmg;
        }
    
        int getX() {
            return x;
        }
    
        void setX(int pX) {
            x = pX;
        }
    
        int getY() {
            return y;
        }
    
        void setY(int pY) {
            y = pY;
        }
};
```

```c++
Player player = Player(12, 2);


std::string key = canvas.lastKey();

while(key != "p") {
    key = canvas.lastKey();
    player.keyMovement(key);
}
```
