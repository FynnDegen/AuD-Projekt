```c++
class Player {
    private:
        int hp;
        int attDmg;
        int attSpd;
    
        int x;
        int y;
        int direc;
        int mapPosX;
        int mapPosY;
    
        Image playerUp;
        Image playerLeft;
        Image playerDown;
        Image playerRight;
        Sword sword;
        SVG *canvas;
    
    public:
        Player(int pMapX, int pMapY, Tile tileMap[15][15] , SVG *canvas){
            mapPosX = pMapX;
            mapPosY = pMapY;
            this -> canvas = canvas
            sword = Sword(1,1, tileMap, canvas);
            playerUp = Image("PlayerUp.png", tileMap[mapPosY][mapPosX].getX(), tileMap[mapPosY][mapPosX].getY(), 100, 100, canvas);
            playerLeft = Image("PlayerLeft.png", tileMap[mapPosY][mapPosX].getX(), tileMap[mapPosY][mapPosX].getY(),100,100, canvas);
            playerDown = Image("PlayerDownTest.png", tileMap[mapPosY][mapPosX].getX(), tileMap[mapPosY][mapPosX].getY(), 100, 100, canvas);
            playerRight = Image("PlayerRight.png", tileMap[mapPosY][mapPosX].getX(), tileMap[mapPosY][mapPosX].getY(), 100, 100, canvas);
        }
    
        void movePlayerX() {
            setX(test[mapPosX][mapPosY].getMapX());
            movePlayerSprite();
        }
    
        void movePlayerY() {
            setY(test[mapPosX][mapPosY].getMapY());
            movePlayerSprite();
        }
    
        void movePlayerSprite() {
            playerUp.moveTo(tileMap[mapPosY][mapPosX].getX(), tileMap[mapPosY][mapPosX].getY());
            playerUp.moveTo(tileMap[mapPosY][mapPosX].getX(), tileMap[mapPosY][mapPosX].getY());
            playerUp.moveTo(tileMap[mapPosY][mapPosX].getX(), tileMap[mapPosY][mapPosX].getY());
            playerUp.moveTo(tileMap[mapPosY][mapPosX].getX(), tileMap[mapPosY][mapPosX].getY());
        }

        void keyMovement(std::string pKey) {
            if(pKey == "w") {
                playerUp.toFront();
                direc = 0;
                if(test[mapPosY - 1][mapPosX].getTag() == 0){
                    mapPosY--;
                    movePlayerY();
                    sword.moveSword(mapPosX, mapPosY, 0);
                } else {
                    sword.moveSword(mapPosX, mapPosY, 0);
                }
            } else if(pKey == "a") {
                playerLeft.toFront();
                direc = 1;
                if(tileMap[mapPosY][mapPosX - 1].getTag() == 0){
                    mapPosX--;
                    movePlayerX();
                    sword.moveSword(mapPosX, mapPosY, 1);
                } else {
                    sword.moveSword(mapPosX, mapPosY, 1);
                }
            } else if(pKey == "s") {
                playerDown.toFront();
                direc = 2;
                if(tileMap[mapPosY + 1][mapPosX].getTag() == 0){
                    mapPosY++;
                    movePlayerY();
                    sword.moveSword(mapPosX, mapPosY, 2);
                } else {
                    sword.moveSword(mapPosX, mapPosY, 2);
                }
            } else if(pKey == "d") {
                playerRight.toFront();
                direc = 3;
                if(tileMap[mapPosY][mapPosX + 1].getTag() == 0){
                    mapPosX++;
                    movePlayerX();
                    sword.moveSword(mapPosX, mapPosY, 3);
                } else {
                    sword.moveSword(mapPosX, mapPosY, 3);
                }

            } else if(pKey == "q") {
                sword.melee();
            } else if(pKey == "ArrowUp") {
                playerUp.toFront();
                direc = 0;
                sword.moveSword(mapPosX, mapPosY, 0);
            } else if(pKey == "ArrowLeft") {
                playerLeft.toFront();
                direc = 1;
                sword.moveSword(mapPosX, mapPosY, 1);
            } else if(pKey == "ArrowDown") {
                playerDown.toFront();
                direc = 2;
                sword.moveSword(mapPosX, mapPosY, 2);
            } else if(pKey == "ArrowRight") {
                playerRight.toFront();
                direc = 3;
                sword.moveSword(mapPosX, mapPosY, 3);
            }
        }
    
        int getHP() {
            return hp;
        }

        void setHP(int pHP) {
            hp = pHP;
        }

        int getAttDmg() {
            return attDmg;
        }

        void setAttDmg(int pAttDmg) {
            attDmg = pAttDmg;
        }
    
        int getX() {
            return x;
        }
    
        void setX(int pX) {
            x = pX;
        }
    
        int getY() {
            return y;
        }
    
        void setY(int pY) {
            y = pY;
        }
};
```
